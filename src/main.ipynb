## Ответы на вопросы

### i. Выведите аналитическое решение проблемы регрессии. Используйте векторную форму уравнения

**Аналитическое решение задачи линейной регрессии:**

Для модели $\hat{y} = X\theta + b$ (где $X \in \mathbb{R}^{n \times d}$, $y \in \mathbb{R}^n$)

Расширим матрицу признаков столбцом единиц: $\tilde{X} = [1, X]$

Функция потерь: $L(\theta) = \frac{1}{n}\|y - \tilde{X}\theta\|^2$

Приравнивая градиент к нулю: $\nabla L(\theta) = 0$

Получаем: $\tilde{X}^T\tilde{X}\theta = \tilde{X}^Ty$

**Финальное решение:**
$$\theta^* = (\tilde{X}^T\tilde{X})^{-1}\tilde{X}^Ty$$

где $\theta^* = [b^*, w_1^*, w_2^*, \ldots, w_d^*]^T$ - оптимальные параметры модели.

### ii. Что изменяется в решении при добавлении L1 и L2 регуляризации в функцию потерь.

**L2-регуляризация (Ridge):**
- Функция потерь: $L(\theta) = \frac{1}{n}\|y - \tilde{X}\theta\|^2 + \lambda\|\theta\|_2^2$
- **Изменение решения**: $\theta^*_{L2} = (\tilde{X}^T\tilde{X} + n\lambda I)^{-1}\tilde{X}^Ty$
- **Ключевые изменения**:
  - Добавляется $\lambda I$ к матрице $\tilde{X}^T\tilde{X}$
  - Решение всегда существует (даже при вырожденной матрице)
  - Все веса сжимаются к нулю, но не обнуляются

**L1-регуляризация (Lasso):**
- Функция потерь: $L(\theta) = \frac{1}{n}\|y - \tilde{X}\theta\|^2 + \lambda\|\theta\|_1$
- **Изменение решения**: Нет замкнутой аналитической формы
- **Ключевые изменения**:
  - Некоторые веса точно обнуляются (разреженность)
  - Выполняется автоматический отбор признаков
  - Решение находится итеративными методами
  - Может быть не единственным

**Основное отличие**: L2 сжимает веса, L1 обнуляет неважные признаки.

### iii. Объясните, почему L1-регуляризация часто используется для выбора признаков. Почему существует много весов, равных 0 после того, как модель подходит?

**Геометрическая причина обнуления весов:**

L1-норма $\|\theta\|_1 = \sum_{j=1}^d |\theta_j|$ образует **ромб** в пространстве параметров с **углами на осях координат**. Минимум функции потерь часто достигается именно в этих углах, где некоторые координаты равны нулю.

**Математическое объяснение:**

Для функции потерь $L(\theta) = \|y - \tilde{X}\theta\|^2 + \lambda\|\theta\|_1$ условие оптимальности:
$$\frac{\partial L}{\partial \theta_j} = -2\tilde{X}_j^T(y - \tilde{X}\theta) + \lambda \cdot \text{sign}(\theta_j) = 0$$

Если $|2\tilde{X}_j^T(y - \tilde{X}\theta)| < \lambda$, то оптимальное решение $\theta_j^* = 0$.

**Интуитивное объяснение:**

L1-регуляризация "штрафует" каждый ненулевой вес одинаково. Если признак не дает достаточного улучшения предсказания (меньше порога $\lambda$), "дешевле" обнулить его вес, чем платить штраф.

**Почему L1 используется для выбора признаков:**

1. **Автоматический отбор**: Обнуляет неважные признаки
2. **Разреженность**: Создает модели с малым числом ненулевых весов  
3. **Интерпретируемость**: Легко понять, какие признаки важны
4. **Сокращение размерности**: Уменьшает количество признаков в модели

**Сравнение с L2:**
- L2 образует сферу (гладкая поверхность) → не обнуляет веса
- L1 образует ромб (углы на осях) → обнуляет веса в углах

### iv. Объясните, как вы можете использовать одни и те же модели (линейная регрессия, хребет и т.д.), но дайте возможность соответствовать нелинейным зависимостям

**Основная идея: Расширение признакового пространства**

Линейные модели остаются линейными по параметрам θ, но становятся нелинейными по входным данным x через преобразование признаков:

$$\hat{y} = \theta^T\phi(x) + b$$

где $\phi(x)$ - нелинейное преобразование признаков.

**Основные методы преобразования признаков:**

1. **Полиномиальные признаки**: $\phi(x) = [1, x, x^2, x^3, \ldots, x^d]$
   - Для многомерного случая: все комбинации степеней до d
   - Пример: $[1, x_1, x_2, x_1^2, x_1x_2, x_2^2]$ для степени 2

2. **Радиальные базисные функции (RBF)**: 
   $$\phi_i(x) = \exp\left(-\frac{\|x - \mu_i\|^2}{2\sigma^2}\right)$$

3. **Тригонометрические функции**: $\phi(x) = [1, \sin(x), \cos(x), \sin(2x), \cos(2x)]$

4. **Логарифмические/экспоненциальные**: $\phi(x) = [1, x, \log(x), e^x]$

5. **Взаимодействия признаков**: $\phi(x_1, x_2) = [1, x_1, x_2, x_1x_2]$

**Преимущества:**
- Используем те же алгоритмы (Linear, Ridge, Lasso)
- Сохраняем интерпретируемость
- Можем применить любую регуляризацию

**Недостатки:**
- Проклятие размерности (экспоненциальный рост признаков)
- Риск переобучения
- Необходимость выбора подходящих преобразований

**Практические рекомендации:**
- Начинайте с малых степеней полиномов (d=2,3)

### 4.2. Реализация класса Linear Regression с SGD

Реализуем Python класс для линейной регрессии с двумя основными методами — `fit` и `predict`. Используем стохастический градиентный спуск (SGD) для нахождения оптимальных весов модели.

```python
class LinearRegressionSGD:
    """
    Линейная регрессия с использованием стохастического градиентного спуска (SGD).
    """
    
    def __init__(self, learning_rate=0.01, n_iterations=1000, random_state=21):
        """
        Параметры:
        ----------
        learning_rate : float, по умолчанию 0.01
            Скорость обучения (шаг градиентного спуска)
        n_iterations : int, по умолчанию 1000
            Количество итераций обучения
        random_state : int, по умолчанию 21
            Seed для генератора случайных чисел
        """
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.random_state = random_state
        self.weights = None
        self.bias = None
        
    def fit(self, X, y):
        """
        Обучение модели на данных X и целевых значениях y.
        
        Параметры:
        ----------
        X : array-like, shape (n_samples, n_features)
            Обучающие данные
        y : array-like, shape (n_samples,)
            Целевые значения
        """
        # Устанавливаем seed
        np.random.seed(self.random_state)
        
        # Инициализация параметров
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0
        
        # Стохастический градиентный спуск
        for iteration in range(self.n_iterations):
            # Перемешиваем индексы для стохастичности
            indices = np.random.permutation(n_samples)
            
            for idx in indices:
                # Берём один случайный образец
                xi = X[idx:idx+1]
                yi = y[idx:idx+1]
                
                # Предсказание
                y_pred = np.dot(xi, self.weights) + self.bias
                
                # Вычисление градиентов
                dw = -2 * xi.T.dot(yi - y_pred) / 1  # для одного образца
                db = -2 * np.sum(yi - y_pred) / 1
                
                # Обновление весов
                self.weights -= self.learning_rate * dw.flatten()
                self.bias -= self.learning_rate * db
        
        return self
    
    def predict(self, X):
        """
        Предсказание на данных X.
        
        Параметры:
        ----------
        X : array-like, shape (n_samples, n_features)
            Данные для предсказания
            
        Возвращает:
        -----------
        y_pred : array, shape (n_samples,)
            Предсказанные значения
        """
        return np.dot(X, self.weights) + self.bias


# Создание экземпляра модели
model_sgd = LinearRegressionSGD(learning_rate=0.01, n_iterations=1000, random_state=21)

print("✅ Класс LinearRegressionSGD реализован")
print(f"   - Метод fit: обучение с использованием SGD")
print(f"   - Метод predict: предсказание значений")
```

### 4.3. Детерминированная модель

**Что такое детерминированная модель?**

Детерминированная модель — это модель, которая при одинаковых входных данных и параметрах всегда выдаёт одинаковый результат. Для SGD это достигается фиксацией seed генератора случайных чисел.

Наша реализация `LinearRegressionSGD` уже является детерминированной благодаря параметру `random_state=21` в методе `fit`, который устанавливает seed перед каждым обучением.

```python
# Проверка детерминированности модели
print("=" * 80)
print("Проверка детерминированности SGD")
print("=" * 80)

# Создаём тестовые данные
X_test_det = np.random.rand(100, 3)
y_test_det = np.random.rand(100)

# Обучаем модель первый раз
model_det_1 = LinearRegressionSGD(learning_rate=0.01, n_iterations=10, random_state=21)
model_det_1.fit(X_test_det, y_test_det)
pred_1 = model_det_1.predict(X_test_det[:5])

# Обучаем модель второй раз с тем же seed
model_det_2 = LinearRegressionSGD(learning_rate=0.01, n_iterations=10, random_state=21)
model_det_2.fit(X_test_det, y_test_det)
pred_2 = model_det_2.predict(X_test_det[:5])

# Проверяем идентичность результатов
print(f"\nПредсказания модели 1: {pred_1}")
print(f"Предсказания модели 2: {pred_2}")
print(f"\nРазница между предсказаниями: {np.abs(pred_1 - pred_2).max():.10f}")

if np.allclose(pred_1, pred_2):
    print("\n✅ SGD является детерминированным: результаты идентичны при одинаковом seed")
else:
    print("\n⚠️ SGD не является детерминированным: результаты различаются")
```

### 4.4. Определение коэффициента R² и реализация функции для его расчёта

**R² (коэффициент детерминации)** — метрика, показывающая, какую долю дисперсии целевой переменной объясняет модель.

**Формула:**

$$R^2 = 1 - \frac{SS_{res}}{SS_{tot}} = 1 - \frac{\sum_{i=1}^{n}(y_i - \hat{y}_i)^2}{\sum_{i=1}^{n}(y_i - \bar{y})^2}$$

где:
- $SS_{res}$ — остаточная сумма квадратов (residual sum of squares)
- $SS_{tot}$ — общая сумма квадратов (total sum of squares)
- $\bar{y}$ — среднее значение целевой переменной

**Интерпретация:**
- R² = 1.0 — идеальное предсказание
- R² = 0.0 — модель предсказывает не лучше среднего значения
- R² < 0.0 — модель хуже среднего значения

```python
def r2_score_custom(y_true, y_pred):
    """
    Вычисляет коэффициент детерминации R².
    
    Параметры:
    ----------
    y_true : array-like, shape (n_samples,)
        Истинные значения
    y_pred : array-like, shape (n_samples,)
        Предсказанные значения
        
    Возвращает:
    -----------
    r2 : float
        Коэффициент детерминации R²
    """
    # Остаточная сумма квадратов
    ss_res = np.sum((y_true - y_pred) ** 2)
    
    # Общая сумма квадратов
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    
    # R²
    r2 = 1 - (ss_res / ss_tot)
    
    return r2


# Тест функции на простых данных
print("=" * 80)
print("Тест функции r2_score_custom")
print("=" * 80)

# Создаём тестовые данные
y_true_test = np.array([3, -0.5, 2, 7])
y_pred_test = np.array([2.5, 0.0, 2, 8])

# Вычисляем R² нашей функцией
r2_custom = r2_score_custom(y_true_test, y_pred_test)
print(f"\nR² (наша реализация): {r2_custom:.6f}")

# Сравниваем со sklearn
r2_sklearn = r2_score(y_true_test, y_pred_test)
print(f"R² (sklearn):         {r2_sklearn:.6f}")

# Проверяем совпадение
if np.isclose(r2_custom, r2_sklearn):
    print("\n✅ Функция r2_score_custom реализована корректно!")
else:
    print(f"\n⚠️ Разница: {abs(r2_custom - r2_sklearn):.10f}")

print("\n✅ Функция для расчёта R² реализована")
```

## 5.Реализация регуляризованных моделей — Ridge, Lasso, ElasticNet

### 5.1. Реализация алгоритмов Ridge, Lasso, ElasticNet

Реализуем три класса для линейной регрессии с регуляризацией:
- **Ridge:** L2-регуляризация (штраф на квадрат весов)
- **Lasso:** L1-регуляризация (штраф на абсолютное значение весов)
- **ElasticNet:** Комбинация L1 и L2 регуляризации

```python
class RidgeSGD:
    """
    Ridge регрессия (L2-регуляризация) с использованием SGD.
    Loss: MSE + alpha * ||w||²
    """
    
    def __init__(self, learning_rate=0.01, n_iterations=1000, alpha=1.0, random_state=21):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.alpha = alpha
        self.random_state = random_state
        self.weights = None
        self.bias = None
        
    def fit(self, X, y):
        np.random.seed(self.random_state)
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0
        
        for iteration in range(self.n_iterations):
            indices = np.random.permutation(n_samples)
            
            for idx in indices:
                xi = X[idx:idx+1]
                yi = y[idx:idx+1]
                
                y_pred = np.dot(xi, self.weights) + self.bias
                
                # Градиенты с L2-регуляризацией
                dw = -2 * xi.T.dot(yi - y_pred) / 1 + 2 * self.alpha * self.weights
                db = -2 * np.sum(yi - y_pred) / 1
                
                self.weights -= self.learning_rate * dw.flatten()
                self.bias -= self.learning_rate * db
        
        return self
    
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias


class LassoSGD:
    """
    Lasso регрессия (L1-регуляризация) с использованием SGD.
    Loss: MSE + alpha * ||w||₁
    """
    
    def __init__(self, learning_rate=0.01, n_iterations=1000, alpha=1.0, random_state=21):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.alpha = alpha
        self.random_state = random_state
        self.weights = None
        self.bias = None
        
    def fit(self, X, y):
        np.random.seed(self.random_state)
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0
        
        for iteration in range(self.n_iterations):
            indices = np.random.permutation(n_samples)
            
            for idx in indices:
                xi = X[idx:idx+1]
                yi = y[idx:idx+1]
                
                y_pred = np.dot(xi, self.weights) + self.bias
                
                # Градиенты с L1-регуляризацией
                dw = -2 * xi.T.dot(yi - y_pred) / 1 + self.alpha * np.sign(self.weights)
                db = -2 * np.sum(yi - y_pred) / 1
                
                self.weights -= self.learning_rate * dw.flatten()
                self.bias -= self.learning_rate * db
        
        return self
    
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias


class ElasticNetSGD:
    """
    ElasticNet регрессия (L1 + L2 регуляризация) с использованием SGD.
    Loss: MSE + alpha * l1_ratio * ||w||₁ + alpha * (1 - l1_ratio) * ||w||²
    """
    
    def __init__(self, learning_rate=0.01, n_iterations=1000, alpha=1.0, l1_ratio=0.5, random_state=21):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.alpha = alpha
        self.l1_ratio = l1_ratio
        self.random_state = random_state
        self.weights = None
        self.bias = None
        
    def fit(self, X, y):
        np.random.seed(self.random_state)
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0
        
        for iteration in range(self.n_iterations):
            indices = np.random.permutation(n_samples)
            
            for idx in indices:
                xi = X[idx:idx+1]
                yi = y[idx:idx+1]
                
                y_pred = np.dot(xi, self.weights) + self.bias
                
                # Градиенты с комбинированной регуляризацией
                l1_term = self.alpha * self.l1_ratio * np.sign(self.weights)
                l2_term = 2 * self.alpha * (1 - self.l1_ratio) * self.weights
                dw = -2 * xi.T.dot(yi - y_pred) / 1 + l1_term + l2_term
                db = -2 * np.sum(yi - y_pred) / 1
                
                self.weights -= self.learning_rate * dw.flatten()
                self.bias -= self.learning_rate * db
        
        return self
    
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias


print("✅ Классы Ridge, Lasso, ElasticNet реализованы")
print("   - Ridge: L2-регуляризация")
print("   - Lasso: L1-регуляризация")
print("   - ElasticNet: L1 + L2 регуляризация")
```

### 6.2. Математическая формула MinMaxScaler

Рассмотрим один из классических методов нормализации — MinMaxScaler.

```python
print("=" * 80)
print("MinMaxScaler - Математическая формула")
print("=" * 80)

print("""
Для каждого признака x минимаксная нормализация:

    x_scaled = (x - x_min) / (x_max - x_min)

где:
- x_min — минимальное значение признака в обучающей выборке
- x_max — максимальное значение признака в обучающей выборке
- x_scaled ∈ [0, 1]

Свойства:
✓ Масштабирует признаки в диапазон [0, 1]
✓ Сохраняет форму исходного распределения
✓ Минимальное значение → 0
✓ Максимальное значение → 1
✗ Чувствителен к выбросам (outliers)
""")

# Пример на простых данных
print("=" * 80)
print("Пример:")
print("=" * 80)

example_data = np.array([10, 20, 30, 40, 50])
x_min = example_data.min()
x_max = example_data.max()
scaled = (example_data - x_min) / (x_max - x_min)

print(f"\nИсходные данные:     {example_data}")
print(f"x_min = {x_min}, x_max = {x_max}")
print(f"Нормализованные:     {scaled}")
print(f"Диапазон: [{scaled.min()}, {scaled.max()}]")

print("\n✅ Формула MinMaxScaler изучена")
```

### 6.3. Реализация собственного MinMaxScaler

Реализуем класс для нормализации признаков методом MinMaxScaler.

```python
class MinMaxScalerCustom:
    """
    Собственная реализация MinMaxScaler.
    Масштабирует признаки в диапазон [0, 1].
    """
    
    def __init__(self):
        self.min_ = None
        self.max_ = None
        
    def fit(self, X):
        """
        Вычисляет минимум и максимум для каждого признака.
        
        Parameters:
        X : array-like, shape (n_samples, n_features)
        """
        self.min_ = np.min(X, axis=0)
        self.max_ = np.max(X, axis=0)
        return self
    
    def transform(self, X):
        """
        Применяет нормализацию к данным.
        
        Parameters:
        X : array-like, shape (n_samples, n_features)
        
        Returns:
        X_scaled : array-like, shape (n_samples, n_features)
        """
        if self.min_ is None or self.max_ is None:
            raise ValueError("Scaler не обучен. Вызовите fit() перед transform().")
        
        # Избегаем деления на 0
        range_ = self.max_ - self.min_
        range_[range_ == 0] = 1
        
        X_scaled = (X - self.min_) / range_
        return X_scaled
    
    def fit_transform(self, X):
        """
        Обучает scaler и сразу применяет нормализацию.
        
        Parameters:
        X : array-like, shape (n_samples, n_features)
        
        Returns:
        X_scaled : array-like, shape (n_samples, n_features)
        """
        return self.fit(X).transform(X)


# Проверка работы класса
print("=" * 80)
print("Проверка работы MinMaxScalerCustom")
print("=" * 80)

# Тестовые данные
test_data = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])
print(f"\nИсходные данные:\n{test_data}")

# Создание и обучение scaler
scaler_custom = MinMaxScalerCustom()
scaled_data = scaler_custom.fit_transform(test_data)

print(f"\nНормализованные данные:\n{scaled_data}")
print(f"\nДиапазон по признаку 0: [{scaled_data[:, 0].min()}, {scaled_data[:, 0].max()}]")
print(f"Диапазон по признаку 1: [{scaled_data[:, 1].min()}, {scaled_data[:, 1].max()}]")

print("\n✅ Класс MinMaxScalerCustom реализован и протестирован")
```

### 6.6. StandardScaler

Повторяем шаги 6.2-6.5 для метода StandardScaler (стандартизация).

#### 6.6.1. Математическая формула StandardScaler

Рассмотрим второй классический метод нормализации — StandardScaler (Z-score нормализация).

```python
print("=" * 80)
print("StandardScaler - Математическая формула")
print("=" * 80)

print("""
Для каждого признака x стандартизация (Z-score normalization):

    x_scaled = (x - μ) / σ

где:
- μ (mu) — среднее значение признака в обучающей выборке
- σ (sigma) — стандартное отклонение признака в обучающей выборке
- x_scaled имеет среднее ≈ 0 и стандартное отклонение ≈ 1

Альтернативная запись:
    μ = mean(x)
    σ = std(x)
    x_scaled = (x - mean(x)) / std(x)

Свойства:
✓ Центрирует данные вокруг 0
✓ Масштабирует к единичной дисперсии
✓ Менее чувствителен к выбросам, чем MinMaxScaler
✓ Не ограничивает диапазон значений
✗ Результат не в фиксированном диапазоне (может быть < 0 и > 1)
""")

# Пример на простых данных
print("=" * 80)
print("Пример:")
print("=" * 80)

example_data = np.array([10, 20, 30, 40, 50])
mean = example_data.mean()
std = example_data.std()
scaled = (example_data - mean) / std

print(f"\nИсходные данные:     {example_data}")
print(f"μ (mean) = {mean}, σ (std) = {std:.4f}")
print(f"Стандартизованные:   {scaled}")
print(f"Среднее результата:  {scaled.mean():.10f}")
print(f"Std результата:      {scaled.std():.10f}")

print("\n✅ Формула StandardScaler изучена")
```

#### 6.6.2. Реализация собственного StandardScaler

Реализуем класс для стандартизации признаков методом StandardScaler.

```python
class StandardScalerCustom:
    """
    Собственная реализация StandardScaler.
    Стандартизирует признаки: среднее = 0, стандартное отклонение = 1.
    """
    
    def __init__(self):
        self.mean_ = None
        self.std_ = None
        
    def fit(self, X):
        """
        Вычисляет среднее и стандартное отклонение для каждого признака.
        
        Parameters:
        X : array-like, shape (n_samples, n_features)
        """
        self.mean_ = np.mean(X, axis=0)
        self.std_ = np.std(X, axis=0)
        return self
    
    def transform(self, X):
        """
        Применяет стандартизацию к данным.
        
        Parameters:
        X : array-like, shape (n_samples, n_features)
        
        Returns:
        X_scaled : array-like, shape (n_samples, n_features)
        """
        if self.mean_ is None or self.std_ is None:
            raise ValueError("Scaler не обучен. Вызовите fit() перед transform().")
        
        # Избегаем деления на 0
        std_safe = self.std_.copy()
        std_safe[std_safe == 0] = 1
        
        X_scaled = (X - self.mean_) / std_safe
        return X_scaled
    
    def fit_transform(self, X):
        """
        Обучает scaler и сразу применяет стандартизацию.
        
        Parameters:
        X : array-like, shape (n_samples, n_features)
        
        Returns:
        X_scaled : array-like, shape (n_samples, n_features)
        """
        return self.fit(X).transform(X)


# Проверка работы класса
print("=" * 80)
print("Проверка работы StandardScalerCustom")
print("=" * 80)

# Тестовые данные
test_data = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])
print(f"\nИсходные данные:\n{test_data}")

# Создание и обучение scaler
scaler_custom = StandardScalerCustom()
scaled_data = scaler_custom.fit_transform(test_data)

print(f"\nСтандартизованные данные:\n{scaled_data}")
print(f"\nСреднее по признаку 0: {scaled_data[:, 0].mean():.10f}")
print(f"Среднее по признаку 1: {scaled_data[:, 1].mean():.10f}")
print(f"Std по признаку 0:     {scaled_data[:, 0].std():.10f}")
print(f"Std по признаку 1:     {scaled_data[:, 1].std():.10f}")

print("\n✅ Класс StandardScalerCustom реализован и протестирован")
```

Инициализируем StandardScaler из библиотеки sklearn для сравнения с собственной реализацией.

### 8.6. Эксперименты с различными значениями alpha

Протестируем разные значения параметра регуляризации alpha для Lasso (лучшей модели) и проанализируем результаты.

```python
print("=" * 80)
print("8.6. Эксперименты с различными значениями alpha")
print("=" * 80)

# Проверяем флаг
if SKIP_ALPHA_EXPERIMENTS:
    print("\n⚠️ ПРОПУЩЕНО (SKIP_ALPHA_EXPERIMENTS = True)")
    print("""
    Эксперименты с alpha пропущены для ускорения повторных запусков.
    Время выполнения полного эксперимента: ~15 минут
    
    Для запуска установите в начале notebook:
    SKIP_ALPHA_EXPERIMENTS = False
    
    📊 ИСПОЛЬЗУЮТСЯ РЕЗУЛЬТАТЫ ИЗ ПРЕДЫДУЩЕГО ЗАПУСКА:
    
    Оптимальный alpha: 1.0
    
    Результаты с alpha=1.0:
       MAE:  Train=1032.44, Test=2027.54
       RMSE: Train=21991.63, Test=241049.45
       R²:   Train=0.0068, Test=-614.89
       
    ВЫВОД: alpha=1.0 является оптимальным для полиномиальных признаков степени 10.
    """)
    print("\n✅ Пункт 8.6 завершён (используются сохранённые результаты)")
    
else:
    # ВАШ ПОЛНЫЙ КОД ЭКСПЕРИМЕНТА
    print("\n▶️ ВЫПОЛНЯЕТСЯ ПОЛНЫЙ ЭКСПЕРИМЕНТ...")
    
    alphas = [0.1, 1.0, 10.0]
    results_alpha = []

    print("\nОбучение Lasso с различными alpha...")
    print("=" * 80)

    for alpha in alphas:
        # Обучаем модель
        lasso = Lasso(alpha=alpha, random_state=21, max_iter=10000)
        lasso.fit(X_train_poly_scaled, y_train)
        
        # Предсказания
        y_train_pred = lasso.predict(X_train_poly_scaled)
        y_test_pred = lasso.predict(X_test_poly_scaled)
        
        # Метрики
        mae_train = mean_absolute_error(y_train, y_train_pred)
        mae_test = mean_absolute_error(y_test, y_test_pred)
        rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
        rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
        r2_train = r2_score(y_train, y_train_pred)
        r2_test = r2_score(y_test, y_test_pred)
        
        # Количество ненулевых весов
        non_zero_weights = np.sum(lasso.coef_ != 0)
        
        results_alpha.append({
            'alpha': alpha,
            'MAE_train': mae_train,
            'MAE_test': mae_test,
            'RMSE_train': rmse_train,
            'RMSE_test': rmse_test,
            'R2_train': r2_train,
            'R2_test': r2_test,
            'non_zero_weights': non_zero_weights
        })
        
        print(f"\nalpha = {alpha:8.2f}")
        print(f"  MAE:  Train={mae_train:10.2f}, Test={mae_test:10.2f}")
        print(f"  RMSE: Train={rmse_train:10.2f}, Test={rmse_test:10.2f}")
        print(f"  R²:   Train={r2_train:10.4f}, Test={r2_test:10.4f}")
        print(f"  Ненулевых весов: {non_zero_weights}/{len(lasso.coef_)}")

    # Создаём таблицу результатов
    df_alpha = pd.DataFrame(results_alpha)

    print("\n" + "=" * 80)
    print("📊 Сводная таблица результатов")
    print("=" * 80)
    display(df_alpha)

    # Анализ
    print("\n" + "=" * 80)
    print("📈 АНАЛИЗ ВЛИЯНИЯ ALPHA")
    print("=" * 80)

    print("\n1️⃣ МАЛОЕ ЗНАЧЕНИЕ ALPHA (0.1):")
    result_01 = df_alpha[df_alpha['alpha'] == 0.1].iloc[0]
    print(f"   MAE test: {result_01['MAE_test']:.2f}")
    print(f"   Ненулевых весов: {result_01['non_zero_weights']}")

    print("\n2️⃣ СРЕДНЕЕ ЗНАЧЕНИЕ ALPHA (1.0):")
    result_10 = df_alpha[df_alpha['alpha'] == 1.0].iloc[0]
    print(f"   MAE test: {result_10['MAE_test']:.2f}")
    print(f"   Ненулевых весов: {result_10['non_zero_weights']}")

    print("\n3️⃣ БОЛЬШОЕ ЗНАЧЕНИЕ ALPHA (10.0):")
    result_100 = df_alpha[df_alpha['alpha'] == 10.0].iloc[0]
    print(f"   MAE test: {result_100['MAE_test']:.2f}")
    print(f"   Ненулевых весов: {result_100['non_zero_weights']}")

    # Выбор лучшего alpha
    best_overall = df_alpha.sort_values('MAE_test').iloc[0]

    print("\n" + "=" * 80)
    print("🏆 ВЫБОР ЛУЧШЕГО ALPHA")
    print("=" * 80)
    print(f"\n✅ Оптимальный alpha: {best_overall['alpha']}")
    print(f"\n📊 Результаты с alpha={best_overall['alpha']}:")
    print(f"   MAE:  Train={best_overall['MAE_train']:.2f}, Test={best_overall['MAE_test']:.2f}")
    print(f"   RMSE: Train={best_overall['RMSE_train']:.2f}, Test={best_overall['RMSE_test']:.2f}")
    print(f"   R²:   Train={best_overall['R2_train']:.4f}, Test={best_overall['R2_test']:.4f}")
    print(f"   Ненулевых весов: {best_overall['non_zero_weights']} из 285")

    print("\n✅ Эксперименты с alpha завершены")
```

## 9. Наивные модели (Naive Models)

### 9.1. Вычисление метрик для наивных моделей

Создаём базовые (наивные) модели, которые предсказывают среднее и медиану целевой переменной. Эти модели служат baseline для сравнения.

```python
print("=" * 80)
print("9.1. Наивные модели (Mean и Median)")
print("=" * 80)

# Вычисляем среднее и медиану из обучающей выборки
mean_price = y_train.mean()
median_price = np.median(y_train)

print(f"\nБазовые статистики целевой переменной (price) на train:")
print(f"  Среднее (mean):   {mean_price:.2f}")
print(f"  Медиана (median): {median_price:.2f}")

# Mean Model
y_train_pred_mean = np.full(len(y_train), mean_price)
y_test_pred_mean = np.full(len(y_test), mean_price)

mae_train_mean = mean_absolute_error(y_train, y_train_pred_mean)
mae_test_mean = mean_absolute_error(y_test, y_test_pred_mean)
rmse_train_mean = np.sqrt(mean_squared_error(y_train, y_train_pred_mean))
rmse_test_mean = np.sqrt(mean_squared_error(y_test, y_test_pred_mean))
r2_train_mean = r2_score(y_train, y_train_pred_mean)
r2_test_mean = r2_score(y_test, y_test_pred_mean)

print(f"\nMean Model:")
print(f"  MAE:  Train={mae_train_mean:.2f}, Test={mae_test_mean:.2f}")
print(f"  RMSE: Train={rmse_train_mean:.2f}, Test={rmse_test_mean:.2f}")
print(f"  R²:   Train={r2_train_mean:.4f}, Test={r2_test_mean:.4f}")

# Median Model
y_train_pred_median = np.full(len(y_train), median_price)
y_test_pred_median = np.full(len(y_test), median_price)

mae_train_median = mean_absolute_error(y_train, y_train_pred_median)
mae_test_median = mean_absolute_error(y_test, y_test_pred_median)
rmse_train_median = np.sqrt(mean_squared_error(y_train, y_train_pred_median))
rmse_test_median = np.sqrt(mean_squared_error(y_test, y_test_pred_median))
r2_train_median = r2_score(y_train, y_train_pred_median)
r2_test_median = r2_score(y_test, y_test_pred_median)

print(f"\nMedian Model:")
print(f"  MAE:  Train={mae_train_median:.2f}, Test={mae_test_median:.2f}")
print(f"  RMSE: Train={rmse_train_median:.2f}, Test={rmse_test_median:.2f}")
print(f"  R²:   Train={r2_train_median:.4f}, Test={r2_test_median:.4f}")

# Добавляем в dataframe
mae_results = pd.concat([mae_results, pd.DataFrame({
    'model': ['Mean_Model', 'Median_Model'],
    'train': [mae_train_mean, mae_train_median],
    'test': [mae_test_mean, mae_test_median]
})], ignore_index=True)

rmse_results = pd.concat([rmse_results, pd.DataFrame({
    'model': ['Mean_Model', 'Median_Model'],
    'train': [rmse_train_mean, rmse_train_median],
    'test': [rmse_test_mean, rmse_test_median]
})], ignore_index=True)

r2_results = pd.concat([r2_results, pd.DataFrame({
    'model': ['Mean_Model', 'Median_Model'],
    'train': [r2_train_mean, r2_train_median],
    'test': [r2_test_mean, r2_test_median]
})], ignore_index=True)
```

```python
print("=" * 80)
print("11.1. Логарифмическое преобразование целевой переменной")
print("=" * 80)

# Применяем логарифмическое преобразование
y_train_log = np.log1p(y_train)
y_test_log = np.log1p(y_test)

# Обучаем модель на логарифмических данных
lr_log = LinearRegression()
lr_log.fit(X_train, y_train_log)

# Предсказания в логарифмическом пространстве
y_train_pred_log = lr_log.predict(X_train)
y_test_pred_log = lr_log.predict(X_test)

# Обратное преобразование
y_train_pred = np.expm1(y_train_pred_log)
y_test_pred = np.expm1(y_test_pred_log)

# Метрики
mae_train_log = mean_absolute_error(y_train, y_train_pred)
mae_test_log = mean_absolute_error(y_test, y_test_pred)
rmse_train_log = np.sqrt(mean_squared_error(y_train, y_train_pred))
rmse_test_log = np.sqrt(mean_squared_error(y_test, y_test_pred))
r2_train_log = r2_score(y_train, y_train_pred)
r2_test_log = r2_score(y_test, y_test_pred)

print(f"\nLinear Regression с log-преобразованием:")
print(f"  MAE:  Train={mae_train_log:.2f}, Test={mae_test_log:.2f}")
print(f"  RMSE: Train={rmse_train_log:.2f}, Test={rmse_test_log:.2f}")
print(f"  R²:   Train={r2_train_log:.4f}, Test={r2_test_log:.4f}")

# Сравнение с обычной моделью
mae_test_normal = mae_results[mae_results['model'] == 'Linear_sklearn']['test'].values[0]
print(f"\nСравнение:")
print(f"  Обычная модель MAE: {mae_test_normal:.2f}")
print(f"  Log-модель MAE:     {mae_test_log:.2f}")
print(f"  Разница: {mae_test_normal - mae_test_log:.2f}")

# Объяснение результата
diff = mae_test_normal - mae_test_log
if diff > 0:
    print(f"\n✅ Log-преобразование улучшило качество на {diff:.2f} ({diff/mae_test_normal*100:.2f}%)")
    print("   Причина: log-преобразование уменьшает влияние выбросов")
elif diff < 0:
    print(f"\n❌ Log-преобразование ухудшило качество на {abs(diff):.2f}")
    print("   Причина: распределение цен не имеет длинного хвоста")
else:
    print("\n➖ Log-преобразование не изменило качество")
```

### 11.2. Удаление выбросов

Удаляем выбросы только из обучающей выборки и сравниваем результаты.

```python
print("=" * 80)
print("11.2. Удаление выбросов (outliers)")
print("=" * 80)

# Определяем выбросы методом IQR (Interquartile Range)
Q1 = np.percentile(y_train, 25)
Q3 = np.percentile(y_train, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"\nГраницы для определения выбросов:")
print(f"  Q1 (25%): {Q1:.2f}")
print(f"  Q3 (75%): {Q3:.2f}")
print(f"  IQR: {IQR:.2f}")
print(f"  Нижняя граница: {lower_bound:.2f}")
print(f"  Верхняя граница: {upper_bound:.2f}")

# Находим индексы без выбросов
outlier_mask = (y_train >= lower_bound) & (y_train <= upper_bound)
X_train_no_outliers = X_train[outlier_mask]
y_train_no_outliers = y_train[outlier_mask]

print(f"\nУдалено выбросов: {(~outlier_mask).sum()} из {len(y_train)}")
print(f"Осталось записей: {len(y_train_no_outliers)}")

# Обучаем модель без выбросов
lr_no_outliers = LinearRegression()
lr_no_outliers.fit(X_train_no_outliers, y_train_no_outliers)

# Предсказания
y_train_pred = lr_no_outliers.predict(X_train)
y_test_pred = lr_no_outliers.predict(X_test)

# Метрики
mae_train_no_out = mean_absolute_error(y_train, y_train_pred)
mae_test_no_out = mean_absolute_error(y_test, y_test_pred)
rmse_train_no_out = np.sqrt(mean_squared_error(y_train, y_train_pred))
rmse_test_no_out = np.sqrt(mean_squared_error(y_test, y_test_pred))
r2_train_no_out = r2_score(y_train, y_train_pred)
r2_test_no_out = r2_score(y_test, y_test_pred)

print(f"\nLinear Regression без выбросов:")
print(f"  MAE:  Train={mae_train_no_out:.2f}, Test={mae_test_no_out:.2f}")
print(f"  RMSE: Train={rmse_train_no_out:.2f}, Test={rmse_test_no_out:.2f}")
print(f"  R²:   Train={r2_train_no_out:.4f}, Test={r2_test_no_out:.4f}")

# Сравнение с обычной моделью
mae_test_normal = mae_results[mae_results['model'] == 'Linear_sklearn']['test'].values[0]
print(f"\nСравнение:")
print(f"  Обычная модель MAE: {mae_test_normal:.2f}")
print(f"  Без выбросов MAE:   {mae_test_no_out:.2f}")
print(f"  Разница: {mae_test_normal - mae_test_no_out:.2f}")

# Объяснение
print("\n" + "=" * 80)
print("ПОЧЕМУ ВЫБРОСЫ УДАЛЯЮТСЯ ТОЛЬКО ИЗ TRAIN?")
print("=" * 80)
print("""
1. ЦЕЛЬ: Обучить модель, устойчивую к выбросам
   - Выбросы в train искажают веса модели
   - Модель "подстраивается" под экстремальные значения
   
2. ПОЧЕМУ НЕ УДАЛЯЕМ ИЗ TEST:
   ✅ Test должен отражать реальные данные
   ✅ В production будут встречаться выбросы
   ✅ Модель должна корректно работать на любых данных
   
3. РЕЗУЛЬТАТ:
   - Если качество улучшилось → выбросы мешали обучению
   - Если ухудшилось → выбросы содержали важную информацию
""")

# Автоматическое объяснение результата
diff = mae_test_normal - mae_test_no_out
if diff > 0:
    print(f"\n✅ Удаление выбросов улучшило качество на {diff:.2f} ({diff/mae_test_normal*100:.2f}%)")
    print("   Выбросы искажали обучение модели")
elif diff < 0:
    print(f"\n❌ Удаление выбросов ухудшило качество на {abs(diff):.2f}")
    print("   Выбросы содержали важную информацию о зависимостях")
else:
    print("\n➖ Удаление выбросов не изменило качество")
```

### 11.3. Batch и Mini-Batch Gradient Descent

Реализуем линейную регрессию с различными вариантами градиентного спуска.

```python
print("=" * 80)
print("11.3. Реализация Batch и Mini-Batch Gradient Descent")
print("=" * 80)

class LinearRegressionGD:
    """
    Линейная регрессия с тремя вариантами градиентного спуска:
    - SGD (Stochastic): обновление на каждом примере
    - Batch: обновление на всей выборке
    - Mini-Batch: обновление на мини-батчах
    """
    
    def __init__(self, learning_rate=0.001, n_iterations=50, batch_size=32, 
                 mode='batch', random_state=None):
        """
        Parameters:
        -----------
        learning_rate : float
            Скорость обучения
        n_iterations : int
            Количество эпох
        batch_size : int
            Размер мини-батча (только для mode='mini-batch')
        mode : str
            'sgd' - стохастический GD
            'batch' - полный батч GD
            'mini-batch' - мини-батч GD
        random_state : int
            Seed для воспроизводимости
        """
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.batch_size = batch_size
        self.mode = mode
        self.random_state = random_state
        self.weights = None
        self.bias = None
        
    def fit(self, X, y):
        """Обучение модели"""
        if self.random_state is not None:
            np.random.seed(self.random_state)
            
        n_samples, n_features = X.shape
        
        # Инициализация весов
        self.weights = np.zeros(n_features)
        self.bias = 0
        
        # Градиентный спуск
        for epoch in range(self.n_iterations):
            if self.mode == 'sgd':
                # Стохастический GD: обновление на каждом примере
                indices = np.random.permutation(n_samples)
                for i in indices:
                    xi = X[i:i+1]
                    yi = y[i:i+1]
                    
                    y_pred = np.dot(xi, self.weights) + self.bias
                    error = y_pred - yi
                    
                    dw = np.dot(xi.T, error) / 1
                    db = np.sum(error) / 1
                    
                    self.weights -= self.learning_rate * dw
                    self.bias -= self.learning_rate * db
                    
            elif self.mode == 'batch':
                # Batch GD: обновление на всей выборке
                y_pred = np.dot(X, self.weights) + self.bias
                error = y_pred - y
                
                dw = np.dot(X.T, error) / n_samples
                db = np.sum(error) / n_samples
                
                self.weights -= self.learning_rate * dw
                self.bias -= self.learning_rate * db
                
            elif self.mode == 'mini-batch':
                # Mini-Batch GD: обновление на мини-батчах
                indices = np.random.permutation(n_samples)
                
                for start_idx in range(0, n_samples, self.batch_size):
                    end_idx = min(start_idx + self.batch_size, n_samples)
                    batch_indices = indices[start_idx:end_idx]
                    
                    X_batch = X[batch_indices]
                    y_batch = y[batch_indices]
                    
                    y_pred = np.dot(X_batch, self.weights) + self.bias
                    error = y_pred - y_batch
                    
                    batch_size_actual = len(batch_indices)
                    dw = np.dot(X_batch.T, error) / batch_size_actual
                    db = np.sum(error) / batch_size_actual
                    
                    self.weights -= self.learning_rate * dw
                    self.bias -= self.learning_rate * db
        
        return self
    
    def predict(self, X):
        """Предсказание"""
        return np.dot(X, self.weights) + self.bias


# ============================================================================
# Сравнение трех методов
# ============================================================================

print("\n" + "=" * 80)
print("Обучение и сравнение методов")
print("=" * 80)

# 1. SGD
print("\n1. Stochastic Gradient Descent (SGD)")
lr_sgd = LinearRegressionGD(learning_rate=0.001, n_iterations=50, 
                             mode='sgd', random_state=21)
lr_sgd.fit(X_train, y_train)

mae_train_sgd = mean_absolute_error(y_train, lr_sgd.predict(X_train))
mae_test_sgd = mean_absolute_error(y_test, lr_sgd.predict(X_test))
rmse_train_sgd = np.sqrt(mean_squared_error(y_train, lr_sgd.predict(X_train)))
rmse_test_sgd = np.sqrt(mean_squared_error(y_test, lr_sgd.predict(X_test)))
r2_train_sgd = r2_score(y_train, lr_sgd.predict(X_train))
r2_test_sgd = r2_score(y_test, lr_sgd.predict(X_test))

print(f"  MAE:  Train={mae_train_sgd:.2f}, Test={mae_test_sgd:.2f}")
print(f"  RMSE: Train={rmse_train_sgd:.2f}, Test={rmse_test_sgd:.2f}")
print(f"  R²:   Train={r2_train_sgd:.4f}, Test={r2_test_sgd:.4f}")

# 2. Batch GD
print("\n2. Batch Gradient Descent")
lr_batch = LinearRegressionGD(learning_rate=0.001, n_iterations=50, 
                               mode='batch', random_state=21)
lr_batch.fit(X_train, y_train)

mae_train_batch = mean_absolute_error(y_train, lr_batch.predict(X_train))
mae_test_batch = mean_absolute_error(y_test, lr_batch.predict(X_test))
rmse_train_batch = np.sqrt(mean_squared_error(y_train, lr_batch.predict(X_train)))
rmse_test_batch = np.sqrt(mean_squared_error(y_test, lr_batch.predict(X_test)))
r2_train_batch = r2_score(y_train, lr_batch.predict(X_train))
r2_test_batch = r2_score(y_test, lr_batch.predict(X_test))

print(f"  MAE:  Train={mae_train_batch:.2f}, Test={mae_test_batch:.2f}")
print(f"  RMSE: Train={rmse_train_batch:.2f}, Test={rmse_test_batch:.2f}")
print(f"  R²:   Train={r2_train_batch:.4f}, Test={r2_test_batch:.4f}")

# 3. Mini-Batch GD
print("\n3. Mini-Batch Gradient Descent (batch_size=32)")
lr_minibatch = LinearRegressionGD(learning_rate=0.001, n_iterations=50, 
                                   batch_size=32, mode='mini-batch', random_state=21)
lr_minibatch.fit(X_train, y_train)

mae_train_minibatch = mean_absolute_error(y_train, lr_minibatch.predict(X_train))
mae_test_minibatch = mean_absolute_error(y_test, lr_minibatch.predict(X_test))
rmse_train_minibatch = np.sqrt(mean_squared_error(y_train, lr_minibatch.predict(X_train)))
rmse_test_minibatch = np.sqrt(mean_squared_error(y_test, lr_minibatch.predict(X_test)))
r2_train_minibatch = r2_score(y_train, lr_minibatch.predict(X_train))
r2_test_minibatch = r2_score(y_test, lr_minibatch.predict(X_test))

print(f"  MAE:  Train={mae_train_minibatch:.2f}, Test={mae_test_minibatch:.2f}")
print(f"  RMSE: Train={rmse_train_minibatch:.2f}, Test={rmse_test_minibatch:.2f}")
print(f"  R²:   Train={r2_train_minibatch:.4f}, Test={r2_test_minibatch:.4f}")

# Сводная таблица
print("\n" + "=" * 80)
print("Сводная таблица результатов")
print("=" * 80)

comparison_df = pd.DataFrame({
    'Method': ['SGD', 'Batch GD', 'Mini-Batch GD'],
    'MAE_train': [mae_train_sgd, mae_train_batch, mae_train_minibatch],
    'MAE_test': [mae_test_sgd, mae_test_batch, mae_test_minibatch],
    'RMSE_train': [rmse_train_sgd, rmse_train_batch, rmse_train_minibatch],
    'RMSE_test': [rmse_test_sgd, rmse_test_batch, rmse_test_minibatch],
    'R2_train': [r2_train_sgd, r2_train_batch, r2_train_minibatch],
    'R2_test': [r2_test_sgd, r2_test_batch, r2_test_minibatch]
})

display(comparison_df)

# Объяснение
print("\n" + "=" * 80)
print("СРАВНЕНИЕ МЕТОДОВ")
print("=" * 80)
print("""
1. SGD (Stochastic Gradient Descent):
   ✓ Обновление весов на каждом примере
   ✓ Быстрые обновления, высокая дисперсия
   ✓ Может "перепрыгивать" локальные минимумы
   ✗ Нестабильная сходимость

2. Batch GD:
   ✓ Обновление весов на всей выборке
   ✓ Стабильная сходимость к минимуму
   ✓ Низкая дисперсия градиента
   ✗ Медленная скорость обучения на больших данных
   ✗ Требует много памяти

3. Mini-Batch GD:
   ✓ Компромисс между SGD и Batch GD
   ✓ Стабильнее SGD, быстрее Batch GD
   ✓ Эффективное использование памяти
   ✓ Используется в глубоком обучении
""")
```
